```python
import numpy as np
from matplotlib import pylab as plt
```

```python
def get_temp_acc_list(output_filename):
    
    temp_acc_list = []
    
    with open(output_filename) as f:
        
        split_line = ['', '', '']
        while (split_line != ['#', 'T', 'A(T)']):
            split_line = f.readline().split()[:3]
        
        split_line = f.readline().split()
        while (len(split_line) == 2):
            temp_acc_list.append([float(string) for string in split_line])
            split_line = f.readline().split()
            
    return np.array(temp_acc_list)
```

```python
opt_list = get_temp_acc_list('output_opt.txt')
```

```python
geo_list = get_temp_acc_list('output_geo.txt')
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
plt.axvline(x=2.269, c='k', linestyle='--', linewidth=1.)
plt.plot(opt_list[:,0], opt_list[:,1], c='b', marker='^', ms=4., label='optimized', linewidth=1.)
plt.plot(geo_list[:,0], geo_list[:,1], c='r', marker='s', ms=4., label='geometric', linewidth=1.)
plt.xlim(0., 10.)
plt.ylim(0., 1.)
plt.xlabel('$T$')
plt.ylabel('$A(T)$')
plt.title('2D FM Ising Model, $L$ = 20')
plt.legend(loc=4, frameon=False)
ax.set_aspect(10)
plt.savefig('fig5_reproduction.png', dpi=400);
```
